```python
#creating a Node class
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# creating linkedlist class
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0 # intializing length to track lenth of linkedlist

    def append(self,new_node):
        new_Node = Node(new_node)
        if self.head is None:
            self.head = new_Node
            self.tail = new_Node
        else:
            self.tail.next = new_Node
            self.tail = new_Node
        self.length +=1

    def traverse(self):
        temp_node = self.head
        result =''
        while temp_node is not None:
            result +=str(temp_node.data)
            if temp_node.next:
                result +='-->'
            temp_node = temp_node.next
        return result

    def insert_at_begin(self,new_node):
        new_node = Node(new_node)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.length +=1

    def insert_at_end(self,new_node):
        new_node = Node(new_node)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length +=1

    def inser_at_mid(self,new_node,index):
        new_node = Node(new_node)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        temp = self.head
        for _ in range(index- 1):
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node
        self.length +=1

    def delete_at_begin(self):
        temp_node = self.head
        if self.length ==1:
            self.head = None
            self.tail = None

        else:
            curr = self.head.next
            self.head = curr
            temp_node.next = None

    def delete_at_end(self):
        temp_node = self.head
        if self.length ==1:
            self.head = None
            self.tail = None
        while temp_node.next.next:
            temp_node = temp_node.next
        temp_node.next = None
        self.tail = temp_node
        self.length -=1

    def delete_at_middle(self,index):
        temp_node = self.head
        if self.length ==1:
            self.head = None
            self.tail = None

        for _ in range(index - 1):
            temp_node = temp_node.next
        curr = temp_node.next
        temp_node.next = temp_node.next.next
        curr.next = None
        self.length -=1

    def search(self, element):
        temp_node = self.head
        while temp_node:
            if temp_node.data == element:
                print('Element is Founded ')
                return
            temp_node = temp_node.next
        print('Element Not  Found')

    def insert_before_node(self,new_node):
        new_node = Node(new_node)
        if self.head is None:
            print('This List is Empty')
        if self.head.next is None:
            new_node.next = self.head
            self.head = new_node


ll = LinkedList()
while True:
  print('1.Append\n2.delete\n3.update\n4.search\n5.traverse')
  choice = int(input('Enter Your Choice'))
  if choice == 1:
    value = int(input('Enter The Value'))
    ll.append(value)
  elif choice == 2:
    value = int(input('Enter The Value'))
    ll.delete(value)
  elif choice == 3:
    value = int(input('Enter The Value'))
    ll.update(value)
  elif choice == 4:
    value = int(input('Enter The Value'))
    ll.search(value)
  elif choice == 5:
    print(ll.traverse())
  else:
    print('Enter Valid Choice')
    break

```
